```python
def test(a,b):
    try:
        r1 = a+b
        r2 = a/b
        r3 = a*b
        return r1, r2, r3
    except Exception as e:
        return e
```

```python
test(1,2)
```

```python
test(5,0)
```

```python
def test(a,b):
    try:
        r1 = a+b
    except:
        print('Error in a+b')
    try:
        r2 = a/b
    except:
        print('Error in r2 a/b')
    try:
        r3 = a*b
    except:
        print('Error in r3 a*b')

    print(r1,r3)
```

```python
test(1,0)
```
