```python
course = input('Write your course name')
if course.lower() == 'python':
    print('Python course is availabel')
elif course.lower() == 'java':
    print('Java course is availabel')
elif course.lower() == 'c':
    print('C course is availabel')
else:
    print(f'Your course {course} is not avilable')
```

```python
marks = int(input('Marks'))
if marks >= 90 and marks <= 100:
    print('A')
elif marks >= 60 and marks < 90:
    print('B')
elif marks >= 35 and marks < 60:
    print("pass")
elif marks >= 0 and marks < 35:
    print('Fail')
else:
    print('Please write marks between 0-100')
```
