```python
import pandas as pd
```

```python
df = pd.read_csv("titanic (1).csv")
```

```python
df.head()
```

```python
def fare_cat(fare):
    if fare >= 300:
        return 'A'
    elif fare >= 200:
        return 'B'
    else:
        return 'C'
```

```python
df['Fare'].apply(fare_cat)
```

```python
df['Fare_cat'] = df['Fare'].apply(fare_cat)
```
