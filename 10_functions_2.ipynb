```python
# Kwargs
```

```python
def test(**kwargs):
    return kwargs
    
```

```python
test(name='Om', Num=4234)
```

```python
def test2(**kwargs):
    return kwargs.values()
```

```python
test2(name='Om', Num=4234)
```

```python
def test3(*args, **kwargs):
    return args, kwargs
```

```python
test3(1,2,3,4,2,12,3,2,4,name='Om', roll = 'data analyst', age = 30)
```

```python
def test4(num1,num2,*args,**kwargs):
    return num1, num2, args, kwargs
```

```python
test4(10,20,30,40,50,60,name='Om', roll = 'data analyst', age = 30)
```
