```python
for i in range(1,11):
    if i % 2 == 0:
        print(i, 'Even')
    else:
        print(i, 'Odd')
```

```python
k = [10,20,15.20,'Hello', [1,2,2.5,'Python'],True]
```

```python
for i in k:
    print(i, type(i))
```

```python
for i in k:
    if type(i) == int:
        print(i)
    elif type(i) == list:
        for data in i:
          if type(data) == int:
              print(data)
```
