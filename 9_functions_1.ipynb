```python
# Fun argus input filter int type data
```

```python
def test5(*args):
    l = []
    for i in args:
        if type(i) == int:
            l.append(i)
    return l
```

```python
test5(2,3.5,22,12.4,"hello", True)
```

```python
def test6(*args):
    o = [] # local var
    for i in args:
        if type(i) == int:
            o.append(i)
        elif type(i) == list:
            for data in i:
                if type(data) == int:
                    o.append(data)
    return o
```

```python
test6(1,2,3,10.2,3.3,'Hello', [100,200,15.50, 'Python'],True,'By')
```
